```python
# Define the base URL and endpoint
url = 'https://api.fiscaldata.treasury.gov/services/api/fiscal_service/v1/accounting/dts/deposits_withdrawals_operating_cash'

# Initialize parameters for pagination and sorting
limit = 100  # Number of rows to fetch per request (assuming the API limits you to 100)
offset = 0   # Start from the first row
total_rows = 1000  # Target number of rows
all_data = []

# Loop to request data in batches of 'limit' until we get the top 1000 rows
while len(all_data) < total_rows:
    # Define the parameters for the API request (including sorting and pagination)
    params = {
        'limit': limit,          # Fetch 100 rows at a time
        'offset': offset,        # Use the offset to get the next set of rows
        'sort': 'date',          # Sort by date (most recent first)
        'order': 'desc'          # Order descending (newest first)
        # Add any other necessary parameters here (like filters)
    }

    # Make the API request
    response = requests.get(url, params=params)
    data = response.json()
    
    # Extract the relevant part of the data
    rows = data.get('results', [])  # Adjust according to the API structure

    # Append the rows to your data list
    all_data.extend(rows)

    # Increment the offset to get the next batch of data
    offset += limit

    # Break if there are no more rows (in case there are fewer than 1000 rows available)
    if not rows:
        break

# Convert the full dataset into a pandas DataFrame
df = pd.DataFrame(all_data[:total_rows])  # Get exactly 1000 rows (in case more were fetched)

# Display the DataFrame
print(df.head())
```
